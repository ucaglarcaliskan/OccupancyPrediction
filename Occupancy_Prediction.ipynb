```python
from sklearn.metrics import confusion_matrix
def confusion_vis(cm):
    f, ax = plt.subplots(figsize=(5, 5))
    corr = cm
    corr = pd.DataFrame(corr,['Occupancy:0','Occupancy:1'])
    sns.heatmap(corr,
    xticklabels=corr.columns.values,
    yticklabels=corr.columns.values,
    annot=True, cmap=plt.get_cmap('Blues'), fmt='g', annot_kws={"size": 16})
    b, t = plt.ylim() # discover the values for bottom and top
    b += 0.5 # Add 0.5 to the bottom
    t -= 0.5 # Subtract 0.5 from the top
    plt.ylim(b, t) # update the ylim(bottom, top) values
    plt.show()
```

### Logistic Regression <a class="anchor" id="log_reg1"></a>

```python
from sklearn.linear_model import LogisticRegression
def logistic_reg(X_train, y_train, X_test, y_test, C):
    logr = LogisticRegression(random_state=42, C=C)
    logr.fit(X_train,y_train) #training

    y_pred = logr.predict(X_test)
    y_pred_t = logr.predict(X_train)
    cm = confusion_matrix(y_test,y_pred)
    confusion_vis(cm)
    
    print('Logistic Regression')
    print('Train Accuracy: %.4f' %logr.score(X_train,y_train))
    print('Test Accuracy: %.4f' %logr.score(X_test,y_test))
    print(cm)
    precision_t = precision_score(y_train, y_pred_t)
    print('Precision for Train Set: %.4f' % precision_t)
    precision = precision_score(y_test, y_pred)
    print('Precision for Test Set: %.4f' % precision)
    recall_t = recall_score(y_train, y_pred_t)
    print('Recall for Train Set %.4f' % recall_t)
    recall = recall_score(y_test, y_pred)
    print('recall for Test Set: %.4f' %recall)
    print('Cross Validation Train Score:' ,cross_val_score(logr,X_train,y_train,cv=3,scoring='accuracy'))
    print('Cross Validation Test Score:' ,cross_val_score(logr,X_test,y_test,cv=3,scoring='accuracy'))
    print('F1 Score of Logistic Regression for Train set')
    f1_t = f1_score(y_train,y_pred_t)
    print(f1_t)
    print('F1 Score of Logistic Regression for Test set')
    f1 = f1_score(y_test,y_pred)
    print(f1)
  
logistic_reg(X_train, y_train, X_test, y_test,1)
```

### KNN <a class="anchor" id="knn1"></a>

#### k=3 <a class="anchor" id="k31"></a>

```python
from sklearn.neighbors import KNeighborsClassifier
def knn_func(X_train, y_train, X_test, y_test,n_neighbors):
    knn = KNeighborsClassifier(n_neighbors=n_neighbors, metric='minkowski')
    knn.fit(X_train,y_train)

    y_pred = knn.predict(X_test)
    y_pred_t = knn.predict(X_train)
    print('kNN')
    cm = confusion_matrix(y_test,y_pred)
    confusion_vis(cm)
    print('Train Accuracy: %.4f' %knn.score(X_train,y_train))
    print('Test Accuracy: %.4f' %knn.score(X_test,y_test))
    print(cm)
    precision_t = precision_score(y_train, y_pred_t)
    print('Precision for Train Set: %.4f' % precision_t)
    precision = precision_score(y_test, y_pred)
    print('Precision for Test Set: %.4f' % precision)
    recall_t = recall_score(y_train, y_pred_t)
    print('Recall for Train Set %.4f' % recall_t)
    recall = recall_score(y_test, y_pred)
    print('recall for Test Set: %.4f' %recall)
    print('Cross Validation Train Score:' ,cross_val_score(knn,X_train,y_train,cv=3,scoring='accuracy'))
    print('Cross Validation Test Score:' ,cross_val_score(knn,X_test,y_test,cv=3,scoring='accuracy'))

    print('F1 Score of kNN for Train set')
    print(f1_score(y_train,y_pred_t))
    print('F1 Score of kNN for Test set')
    print(f1_score(y_test,y_pred))
knn_func(X_train, y_train, X_test, y_test,1)
knn_func(X_train, y_train, X_test, y_test,3)
```

## Decision Tree <a class="anchor" id="dec1"></a>

```python
#Decision tree
from sklearn.tree import DecisionTreeClassifier

def dec_tree(X_train, y_train, X_test, y_test, criterion):
    dtc = DecisionTreeClassifier(criterion = criterion, random_state=42)
    dtc.fit(X_train,y_train)
    y_pred = dtc.predict(X_test)
    y_pred_t = dtc.predict(X_train)
    print('Decision Tree')
    cm = confusion_matrix(y_test,y_pred)
    confusion_vis(cm)
    print('Train Accuracy: %.4f' %dtc.score(X_train,y_train))
    print('Test Accuracy: %.4f' %dtc.score(X_test,y_test))
    print(cm)
    precision = precision_score(y_test, y_pred)
    print('Precision for Test Set: %.4f' % precision)
    recall = recall_score(y_test, y_pred)
    print('recall for Test Set: %.4f' %recall)
    print('F1 Score of Decision Tree for Test set')
    print(f1_score(y_test,y_pred))
dec_tree(X_train, y_train, X_test, y_test,'gini')
dec_tree(X_train, y_train, X_test, y_test,'entropy')
```

```python
def create_isnight(data=None):
    data['isnight'] = 0
    for i in data.index:
        val = data.get_value(i,'hour')
        if val>8 and val<18:
            data.set_value(i,'isnight',0)
        else:
            data.set_value(i,'isnight',1)
    return data
data = create_isnight(data)
test = create_isnight(test)
```
